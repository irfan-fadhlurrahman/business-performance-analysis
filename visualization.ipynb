```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv("Task_3_table.csv")
data.set_index("year")
```

```python
data = pd.read_csv("Task_3_table.csv")
data = data.iloc[1:]

colors_dark = ["#1F1F1F", "#313131", '#636363', '#AEAEAE', '#DADADA']
sns.set_theme(context="talk", style="whitegrid")
fig, ax = plt.subplots(1, 1, figsize=(6, 5),
                                              #dpi=300
)

sns.lineplot(data=data, x='year', y='total_revenue', ax=ax, marker='o', color=colors_dark[1])

for position in ['top', 'left', 'right']:
    ax.spines[position].set_visible(False)
    
ax.axes.yaxis.set_visible(False)
ax.get_xaxis().tick_bottom()

xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()

ax.annotate(
    '9% of total revenue are \nfrom Bed Bath Table \nin 2017', 
    xy=(2017, 6.92*10**6), xytext=(2017.02, 4*10**6),
    fontsize=10,
    fontweight='bold',
    arrowprops=dict(arrowstyle="-",color="0.5",
                            patchB=None,
                            shrinkB=5,
                            connectionstyle="arc,rad=0.1",),
)

ax.annotate(
    'Health Beauty products \ncontribute 10% of total \nrevenue in 2018', 
    xy=(2018, 8.45*10**6), xytext=(2017.57, 2*10**6),
    fontsize=10,
    fontweight='bold',
    arrowprops=dict(arrowstyle="-",color="0.5",
                            patchB=None,
                            shrinkB=5,
                            connectionstyle="arc3,rad=-0.1",),
)

ax.text(
    x=data['year'].iloc[0] - 0.05, 
    y=data['total_revenue'].iloc[0] + 220000, 
    s=round(data['total_revenue'].iloc[0]  / 10**6, 2), 
    fontsize=12,
)

ax.text(
    x=data['year'].iloc[-1] - 0.05, 
    y=data['total_revenue'].iloc[-1] + 210000, 
    s=round(data['total_revenue'].iloc[-1]  / 10**6, 2), 
    fontsize=12,
)
ax.text(
    x=data['year'].iloc[-1] - 0.55, 
    y=data['total_revenue'].iloc[-1] - 400000, 
    s='+22%', 
    fontsize=14,
    fontweight='bold',
    color=colors_dark[3]
)
ax.text(
    x=xmin, 
    y=xmax * 5000, 
    s='Total Revenue per Year', 
    fontsize=20,
    fontweight='bold',
    color=colors_dark[2],
    va='center', 
    ha='left',
)
ax.text(
    x=xmin, 
    y=xmax * 4700, 
    s='# MILLION', 
    fontsize=12,
    color=colors_dark[2],
    va='center', 
    ha='left',
)
ax.text(
    x=xmin + 0.5, 
    y=xmax * 4700, 
    s='% CHANGE', 
    fontsize=12,
    color=colors_dark[2],
    va='center', 
    ha='left',
)

plt.grid(False)
plt.xlabel('')
plt.xticks(np.arange(2017, 2018+1, 1), fontsize=14)
plt.ylim([0, 9000000])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv("task_1_table.csv")
data = data.iloc[1:]

colors_dark = ["#1F1F1F", "#313131", '#636363', '#AEAEAE', '#DADADA']
sns.set_theme(context="talk", style="whitegrid")
fig, ax = plt.subplots(1, 1, figsize=(6, 5),
                       #dpi=300
)

sns.lineplot(data=data, x='year', y='total_new_customer', ax=ax, marker='o', color=colors_dark[1])
sns.lineplot(data=data, x='year', y='avg_monthly_active_user', ax=ax, marker='o', color=colors_dark[1])

for position in ['top', 'left', 'right']:
    ax.spines[position].set_visible(False)

ax.axes.yaxis.set_visible(False)
ax.get_xaxis().tick_bottom()

xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()

ax.text(
    x=data['year'].iloc[-1] + 0.1, 
    y=data['total_new_customer'].iloc[-1] - 0.1, 
    s='New User', 
    fontsize=14,
    va='center', 
    ha='left', 
    #bbox=dict(boxstyle='round,pad=0.3', ec='none', alpha=0.3)
)
ax.text(
    x=data['year'].iloc[-1] + 0.1, 
    y=data['avg_monthly_active_user'].iloc[-1] - 0.1, 
    s='Active User',
    fontsize=14,
    va='center', 
    ha='left', 
    #bbox=dict(boxstyle='round,pad=0.3', ec='none', alpha=0.3)
)
ax.text(
    x=data['year'].iloc[-1] - 0.06, 
    y=data['total_new_customer'].iloc[-1] + xmax*1.1, 
    s=data['total_new_customer'].iloc[-1], 
    fontsize=12
)

ax.text(
    x=data['year'].iloc[0] - 0.05, 
    y=data['total_new_customer'].iloc[0] + xmax*1.1, 
    s=data['total_new_customer'].iloc[0], 
    fontsize=12
)

ax.text(
    x=data['year'].iloc[0] - 0.05, 
    y=data['avg_monthly_active_user'].iloc[0] + xmax*1.1, 
    s=int(data['avg_monthly_active_user'].iloc[0]), 
    fontsize=12,
)

ax.text(
    x=data['year'].iloc[-1] - 0.05, 
    y=data['avg_monthly_active_user'].iloc[-1] + xmax*1.1, 
    s=int(data['avg_monthly_active_user'].iloc[-1]), 
    fontsize=12,
)
ax.text(
    x=data['year'].iloc[-1] - 0.55, 
    y=data['total_new_customer'].iloc[-1] - xmax*1.01, 
    s='+19%', 
    fontsize=14,
    fontweight='bold',
    color=colors_dark[3]
)
ax.text(
    x=data['year'].iloc[-1] - 0.55, 
    y=data['avg_monthly_active_user'].iloc[-1] + 1000, 
    s='+44%', 
    fontsize=14,
    fontweight='bold',
    color=colors_dark[3]
)

ax.text(
    x=xmin, 
    y=xmax*32.1, 
    s='Customer growth by user types', 
    fontsize=20,
    fontweight='bold',
    color=colors_dark[2],
    va='center', 
    ha='left',
)

ax.text(
    x=xmin, 
    y=xmax * 30, 
    s='# OF USERS', 
    fontsize=12,
    color=colors_dark[2],
    va='center', 
    ha='left',
)
ax.text(
    x=xmin + 0.5, 
    y=xmax * 30, 
    s='% CHANGE', 
    fontsize=12,
    color=colors_dark[2],
    va='center', 
    ha='left',
)
plt.grid(False)
plt.xlabel('')
plt.xticks(np.arange(2017, 2018+1, 1), fontsize=14)
plt.ylim([-3000, 60000])
plt.show()
```
